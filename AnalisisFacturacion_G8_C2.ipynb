#1. Análisis de facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna **Precio** de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# Voy a hacer una función que reciba el dataframe y retorne otro DataFrame con la tienda y la facturación total
# Voy a usar Type Hint y Docstrings


def totalFacturacion(df: pd.DataFrame) -> pd.DataFrame:
    '''
    Función que recibe un DataFrame y devuelve otro DataFrame
    con el número de tienda y la sumatoria de lo facturado en cada tienda.

    Columnas de salida:
    - tienda (int)
    - facturacion_total (float)
    '''

    facturacion = df.groupby('tienda', as_index=False)['Precio'].sum()
    facturacion.rename(columns={'Precio': 'facturacion_total'}, inplace=True)
    return facturacion


facturacion_por_tienda = totalFacturacion(df)
print(facturacion_por_tienda)
```

##1.1 Visualizaciones para Análisis de Facturación


*   Bar Plot
*   Pie Chart
*   Line Plot





```python
import matplotlib.pyplot as plt
```

```python
#Bar Plot

# Hacer el Bar Plot directamente al iterar sobre el nuevo DataFrame
plt.figure(figsize=(8,5))
bars = plt.bar(facturacion_por_tienda['tienda'], facturacion_por_tienda['facturacion_total'], color='skyblue')

# Agregar etiquetas arriba de las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5000000, f'{yval:,.0f}', ha='center', va='bottom', fontsize=8)

plt.xlabel('Tienda')
plt.ylabel('Facturación Total en billones')
plt.title('Facturación Total por Tienda')
plt.xticks(facturacion_por_tienda['tienda'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Pie Chart

plt.figure(figsize=(7,4))
plt.pie(facturacion_por_tienda['facturacion_total'],
        labels=None,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors)


# DEFINIMOS las etiquetas antes
labels = [f'Tienda: {tienda}' for tienda in facturacion_por_tienda['tienda']]

# Agregamos la leyenda abajo
plt.legend(
    labels,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.3),
    ncol=2,  # Número de columnas en la leyenda
    frameon=False
)

plt.title('Distribución de Facturación por Tienda')
plt.axis('equal')  # Mantiene el círculo bien redondo
plt.tight_layout()
plt.show()
```

```python
#Line Plot

plt.figure(figsize=(8,5))
plt.plot(facturacion_por_tienda['tienda'], facturacion_por_tienda['facturacion_total'], marker='o', linestyle='-', color='green')

# Agregar valores en cada punto
for i, val in enumerate(facturacion_por_tienda['facturacion_total']):
    plt.text(facturacion_por_tienda['tienda'][i], val + 5000000, f'{val:,.0f}', ha='center', va='bottom', fontsize=8)

plt.xlabel('Tienda')
plt.ylabel('Facturación Total en Billones')
plt.title('Facturación Total por Tienda (Line Plot)')
plt.xticks(facturacion_por_tienda['tienda'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

En este debes calcular la **cantidad de productos vendidos por categoría** en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, **mostrando las categorías más populares de cada tienda**.

```python
# Voy a hacer una función que reciba el dataframe y retorne otro DataFrame con la tienda y la venta por categoría
# Voy a usar Type Hint y Docstrings
import pandas as pd

def catPorTienda(df: pd.DataFrame) -> pd.DataFrame:
    '''
    Función que recibe un DataFrame y devuelve otro DataFrame
    con el número de tienda, la categoría y cantidad de productos de esa categoría vendida.

    Columnas de salida:
    - tienda (int)
    - categoría (str)
    - cantidad (int)
    '''
    categoriasVend = df.groupby(['tienda', 'Categoría del Producto']).size().reset_index(name='cantidad')

    return categoriasVend

# Ejemplo de uso
# Asumiendo que tienes el DataFrame 'df' ya cargado
categoriasVend = catPorTienda(df)
print(categoriasVend)
```

## 2.2 Visualizaciones para análisis por categoría y tienda


*   Pie Chart por cada tienda con % de contribución de cada categoría



```python
import matplotlib.pyplot as plt

# Graficar pie chart para cada tienda
for tienda in categoriasVend['tienda'].unique():
    # Filtrar los datos para la tienda actual
    tienda_data = categoriasVend[categoriasVend['tienda'] == tienda]

    # Crear el gráfico de pie
    plt.figure(figsize=(7,7))
    plt.pie(tienda_data['cantidad'],
            labels=tienda_data['Categoría del Producto'],
            autopct='%1.1f%%',
            startangle=140,
            colors=plt.cm.Paired.colors)

    plt.title(f'Distribución de Productos Vendidos en Tienda {tienda}')
    plt.axis('equal')  # Hace el círculo bien redondo
    plt.show()
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las **calificaciones promedio** de los clientes para cada tienda. El objetivo es conocer la **satisfacción del cliente con los productos vendidos.**

```python
#Investigo el rango min-max de calificaciones del usuario
print(f"La mínima calificación es {df['Calificación'].min()}")
print(f"La máxima calificación es {df['Calificación'].max()}")
```

```python
# Voy a hacer una función que reciba el dataframe y retorne otro DataFrame con la tienda y el promedio de satisfacción del cliente
# Voy a usar Type Hint y Docstrings
def calPromedio(df: pd.DataFrame) -> pd.DataFrame:
    '''
    Función que recibe un DataFrame y devuelve otro DataFrame
    con el número de tienda y el promedio de califacación del usuario.

    Columnas de salida:
    - tienda (int)
    - Calificación Promedio (int64) [1--5]
    '''

    promedio = df.groupby('tienda', as_index=False)['Calificación'].mean()
    promedio.rename(columns={'Calificación': 'calificacion_promedio'}, inplace=True)
    promedio = promedio.sort_values(by='calificacion_promedio', ascending=False)

    return promedio

promedio = calPromedio(df)
print(promedio)
```

## 3.3 Visualización por Satisfacción del cliente

```python
import matplotlib.pyplot as plt
# Crear el bar chart
plt.figure(figsize=(10,6))
bars = plt.bar(promedio['tienda'], promedio['calificacion_promedio'], color='skyblue')

# Agregar etiquetas de valor arriba de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.03, f'{yval:.3f}', ha='center', va='bottom', fontsize=8)

# Etiquetas y título
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio por Tienda (ordenado de mayor a menor)')
plt.xticks(promedio['tienda'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

identificar qué **productos** fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
# Voy a hacer una función que reciba el dataframe y retorne otro DataFrame con la tienda, el producto y la cantidad vendida de cada uno
# El df se ordenará por tienda y por cantidad de productos vendidos
# Voy a usar Type Hint y Docstrings

dfAux = df.copy()
dfAux['CantidadVendidaProducto'] = 1

def prodVendidos(dfAux: pd.DataFrame) -> pd.DataFrame:
    '''
    Función que recibe un DataFrame y devuelve otro DataFrame
    con el número de tienda, producto y la cantidad vendida de cada uno.

    Columnas de salida:
    - tienda (int)
    - Producto (str)
    - Cantidad (int)
    '''

    # Sumar las cantidades vendidas por tienda y producto
    cantProdVendidos = dfAux.groupby(['tienda', 'Producto'], as_index=False)['CantidadVendidaProducto'].sum()
    contProdVendidos = cantProdVendidos.sort_values(by=['tienda', 'CantidadVendidaProducto'], ascending=[True, False])

    return contProdVendidos


prodVendidos_df = prodVendidos(dfAux)

for tienda in prodVendidos_df['tienda'].unique():
    tienda_data = prodVendidos_df[prodVendidos_df['tienda'] == tienda]
    print(f"Tienda {tienda}:")
    print(f'top 5 artículos vendidos en {tienda}: {tienda_data.head(5)}')
    print("\n")
    print(f'menores 5 artículos vendidos en {tienda}: {tienda_data.tail(5)}')
    print("\n")
```

#4.2 Visualizacion de productos mas y menos vendidos por cada tienda

```python
import matplotlib.pyplot as plt
import pandas as pd

# Función de los productos vendidos
def prodVendidos(dfAux: pd.DataFrame) -> pd.DataFrame:
    '''
    Función que recibe un DataFrame y devuelve otro DataFrame
    con el número de tienda, producto y la cantidad vendida de cada uno.

    Columnas de salida:
    - tienda (int)
    - Producto (str)
    - Cantidad (int)
    '''
    # Sumar las cantidades vendidas por tienda y producto
    cantProdVendidos = dfAux.groupby(['tienda', 'Producto'], as_index=False)['CantidadVendidaProducto'].sum()
    contProdVendidos = cantProdVendidos.sort_values(by=['tienda', 'CantidadVendidaProducto'], ascending=[True, False])

    return contProdVendidos

# Crear una copia del DataFrame original
dfAux = df.copy()
dfAux['CantidadVendidaProducto'] = 1

# Llamar la función
prodVendidos_df = prodVendidos(dfAux)

# Configuración para crear el gráfico
fig, axes = plt.subplots(nrows=len(prodVendidos_df['tienda'].unique()), ncols=1, figsize=(10, 6), sharex=True)

for i, tienda in enumerate(prodVendidos_df['tienda'].unique()):
    tienda_data = prodVendidos_df[prodVendidos_df['tienda'] == tienda]

    # Selección de los 5 productos más vendidos
    top_5 = tienda_data.head(5)
    bottom_5 = tienda_data.tail(5)

    # Visualización de los productos más vendidos
    axes[i].bar(top_5['Producto'], top_5['CantidadVendidaProducto'], color='green', label='Top 5 productos vendidos')
    axes[i].bar(bottom_5['Producto'], bottom_5['CantidadVendidaProducto'], color='red', label='Menos vendidos')

    # Etiquetas y título
    axes[i].set_title(f'Tienda {tienda} - Productos Más y Menos Vendidos')
    axes[i].set_ylabel('Cantidad Vendida')
    axes[i].legend()

# Ajuste de las etiquetas y presentación
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
